```python
stemmer=nltk.PorterStemmer()
```

```python
def stemming(list_of_token):
    stemming_word=[]
    
    for word in list_of_token:
        stemming_word.append(stemmer.stem(word))
    
    return stemming_word
```

```python
stemming_data=tokenised_data.apply(lambda x : stemming(x))
```

```python
lemmatizer = nltk.WordNetLemmatizer()
```

```python
def lemmatizing(list_of_word):
    lemmatize_words=[]
    
    for word in list_of_word:
        lemmatize_words.append(lemmatizer.lemmatize(word))
    return lemmatize_words
```

```python
lemmatize_data=tokenised_data.apply(lambda x :lemmatizing(x))
```
